```python
import pandas as pd
```

```python
clients = pd.read_csv("../data/raw/clients.csv")
clients.head()
```

```python
purchases = pd.read_csv(
    "../data/raw/purchases.csv", 
    usecols=["client_id", "transaction_id", "product_id", "product_quantity", "trn_sum_from_iss", "trn_sum_from_red"]
)
purchases.head()
```

```python
products = pd.read_csv("../data/raw/products.csv")
products.head()
```

```python
products["level_1"].nunique(), products["level_2"].nunique(), products["level_3"].nunique(), products["level_4"].nunique()
```

```python
products["segment_id"].nunique()
```

```python
products.set_index("product_id", inplace=True)
```

```python
level_1 = pd.get_dummies(products["level_1"], prefix="level_1")
level_1.head()
```

```python
level_2 = pd.get_dummies(products["level_2"], prefix="level_2")
level_2.head()
```

```python
segment = pd.get_dummies(products["segment_id"].fillna(-1).astype(int), prefix="segment_id")
segment.head()
```

```python
print("level_1")
purchases.join(level_1, on="product_id").sort_values(by="client_id")
```

```python
print("level_2")
purchases.join(level_2, on="product_id").sort_values(by="client_id").to_csv("purchases_level_2.csv", index=True)
print("segment")
purchases.join(segment, on="product_id").sort_values(by="client_id").to_csv("purchases_segment.csv", index=True)
```
